#### 명사(Noun)

```python
# kiwi 명사 추출 함수
kiwi = Kiwi()
def kiwi_noun_extractor(text):
    results = []
    result = kiwi.analyze(text)
    
    for token, pos, _, _ in result[0][0]:
        # kiwi의 태그 목록 체언:
        # NNG: 일반 명사, NNP: 고유 명사, NNB: 의존 명사, NR: 수사, NP: 대명사
        if len(token) != 1 and pos.startswith('N'): #or pos.startswith('SL'):
            results.append(token)
    return results
```

```python
nouns_kiwi = kiwi_noun_extractor(comment_string)
nouns_kiwi
```

```python
# Okt 명사 추출
okt = Okt()
def okt_noun_extractor(text):
    return okt.nouns(text)
        
```

```python
nouns_okt = okt_noun_extractor(comment_string)
nouns_okt
```

#### 형용사(Adjective)

```python
# kiwi 형용사 추출 함수


def kiwi_adj_extractor(text):
    results = []
    result = kiwi.analyze(text)
    
    for token, pos,_,_ in result[0][0]:
        # kiwi의 태그 목록 용언:
        # VA: 형용사
        if len(token) != 1 and pos.startswith('VA'):
            results.append(token)
    return results
```

```python
# Okt 형용사 추출 함수

def okt_adj_extractor(text):
    results = []
    result = okt.pos(text)
    
    for token, pos in result:
        if len(token) != 1 and pos.startswith('Adj'):
            results.append(token)
    return results
```

```python
adj_okt = okt_adj_extractor(comment_string)
adj_okt
```

#### 동사

```python
# kiwi 동사 추출 함수

def kiwi_verb_extractor(text):
    results = []
    result = kiwi.analyze(text)
    
    for token, pos, _,_ in result[0][0]:
        if len(token) != 1 and pos.startswith('VV'):
            results.append(token)
    return results
```

```python
verb_kiwi = kiwi_verb_extractor(comment_string)
verb_kiwi
```

```python
# Okt 동사 추출 함수

def okt_verb_extractor(text):
    results = []
    result = okt.pos(text)
    
    for token, pos in result:
        if pos == 'Verb':
            results.append(token)
    return results
```

```python
verb_okt = okt_verb_extractor(comment_string)
verb_okt
```
