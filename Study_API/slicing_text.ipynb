# 워드 클라우드 시각화

```python
import pandas as pd
```

```python
comment_file = pd.read_csv("C:/Users/pps/Desktop/TIL/youtube_comments_crawling.csv", encoding= 'utf-8')
```

```python
comment_file.drop(columns=['Unnamed: 0'], inplace=True)

```

```python
comment_file=comment_file.fillna(0)
```

```python
# 댓글텍스트를 딕셔너리로 불러오기
comment_file['댓글텍스트'].to_dict()
```

```python
comment_list = comment_file['댓글텍스트'].to_list()
comment_list
```

```python
# iterrows 함수를 사용해 댓글텍스트만 텍스트로 불러오기
comment_slicing_list = []
for index, row in comment_file.iterrows():
    comment_slicing_list.extend(str(row['댓글텍스트']).split(' '))
    
comment_slicing_list
```

```python
from collections import Counter
cnt = Counter()

counts=Counter(comment_slicing_list)
# for i in comment_slicing_list:
#     cnt[i] += 1
# cnt
tags = counts.most_common(30)
print(dict(tags))
```

### Komoran

```python
from konlpy.tag import Komoran
komoran = Komoran(userdic='C:/Users/pps/Desktop/TIL/Study_API/temp_string.txt')
```

```python
# 형태소 분리
# morphs
print(komoran.morphs(u'안녕하세요 만나서 반갑습니다 제이름은 홍길동입니다.'))
print(komoran.morphs(u'샌디에이고의 야구엔 감동이 있다!'))
```

```python
# nouns
print(komoran.nouns(u'아버지를 아버지라 부르지 못하고 형을 형이라 부르지 못한다.'))
print(komoran.nouns(u'샌디에이고의 야구엔 감동이 있다!'))
```

```python
# pos
print(komoran.pos(u'샌디에이고 야구가 너무 재미있다.'))
print(komoran.pos(u'샌디에이고의 야구엔 감동이 있다!'))
```
